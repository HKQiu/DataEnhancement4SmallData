```python
from rdkit import Chem
from rdkit.Chem import Lipinski
import pandas as pd 
import numpy as np
```

```python
def get_num_bond_cycle(smiles):
    mol = Chem.MolFromSmiles(smiles)
    num_bond = Lipinski.NumRotatableBonds(mol)
    num_cycle = Lipinski.NumAliphaticRings(mol) + Lipinski.NumAromaticRings(mol)
    ratio = num_bond / num_cycle
    return num_bond, num_cycle, ratio
```

```python
f = pd.read_csv('D:/0研究生/0聚酰亚胺/高分子Tg数据/加入其他高分子数据的影响/Tg/正式实验/预测筛选/1-350-3_2(Tg-SAscore).csv')
f.head()
```

```python
f['num_bond', 'num_cycle', 'ratio'] = f['Smiles'].apply(get_num_bond_cycle)
```
