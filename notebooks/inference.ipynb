```python
import glob,os
import pandas as pd
import deepchem as dc
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw, PyMol, rdFMCS
from rdkit.Chem.Draw import IPythonConsole
from rdkit import rdBase
from deepchem import metrics
from IPython.display import Image, display
from rdkit.Chem.Draw import SimilarityMaps
import tensorflow as tf

import warnings
warnings.filterwarnings("ignore")
```

```python
DATASET_FILE_Base = '.candidate smiles/'
```

```python
# i-j
res = pd.DataFrame()

for i in range(8):
    for j in range(1,6):
        Val_DATASET_FILE = DATASET_FILE_Base + str(i) + '-' + str(j) + '.csv'
        print(Val_DATASET_FILE)
        Restore_MODEL_DIR = 'Path/To/Model'   # Please replace it with the appropriate path.
        Restore_MODEL_DIR = Restore_MODEL_DIR + '/' + 'loop' + str(2+1)
        print(Restore_MODEL_DIR)
        ########################################Featurizerization#########################
        print("########开始特征化#######")
        featurizer = dc.feat.ConvMolFeaturizer()
        loader = dc.data.CSVLoader(tasks=[], feature_field="Smiles", featurizer=featurizer)
        testset = loader.create_dataset(Val_DATASET_FILE, shard_size=10000)
        print("########特征化完成#######")
        ########################################Model######################################
        model = dc.models.GraphConvModel(1, mode="regression", model_dir=Restore_MODEL_DIR)
        model.restore()
        print("########模型加载完成#######")
        ########################################Predict########################################
        print("########开始预测...#######")
        val_pred = model.predict(testset)
        print("########预测完成，开始保存结果#######")
        pred = pd.DataFrame(val_pred,columns=['pred_new'])

        res = pd.concat([res,pred],axis=0).reset_index(drop=True)
        print("########完成#######")
        print("res.shape:",res.shape)
```

```python

Val_DATASET_FILE = DATASET_FILE_Base + '8.csv'
print(Val_DATASET_FILE)
Restore_MODEL_DIR = 'D:/0研究生/0聚酰亚胺/高分子Tg数据/加入其他高分子数据的影响/Tg/正式实验/模型_选择卷积层/2层'
Restore_MODEL_DIR = Restore_MODEL_DIR + '/' + 'loop' + str(2+1)
print(Restore_MODEL_DIR)
########################################Featurizerization#########################
print("########开始特征化#######")
featurizer = dc.feat.ConvMolFeaturizer()
loader = dc.data.CSVLoader(tasks=[], feature_field="Smiles", featurizer=featurizer)
testset = loader.create_dataset(Val_DATASET_FILE, shard_size=10000)
print("########特征化完成#######")
########################################Model######################################
model = dc.models.GraphConvModel(1, mode="regression", model_dir=Restore_MODEL_DIR)
model.restore()
print("########模型加载完成#######")
########################################Predict########################################
print("########开始预测...#######")
val_pred = model.predict(testset)
print("########预测完成，开始保存结果#######")
pred = pd.DataFrame(val_pred,columns=['pred_new'])

res = pd.concat([res,pred],axis=0).reset_index(drop=True)
print("########所有完成#######")
print("最终结果:",res.shape)
```
